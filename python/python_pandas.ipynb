```python
import numpy as np
import matplotlib.pyplot as plt
```

Importing Data
=======

```python
import pandas as pd
obs=pd.read_csv('ateb_preston/observation_preston.csv' , na_values=['nan'])
```

```python
obs.head()
```

```python
type(obs)
```

```python
type(obs.values)
```

Datetime index & resample
=============

```python
obs.index = pd.date_range(start='20030812-1330',periods=22772,freq='30Min')
obs.head()
```

```python
obs[['QHup','QEup']].resample('15min').interpolate(method='linear')
```

Groupby
==========

```python
obsDJF=obs['2003-12':'2004-02']
```

```python
obsDJF
```

```python
obsDJF['NetRad'].groupby(obsDJF.index.hour).describe()
```

```python
hour_month=obsDJF['NetRad'].groupby([obsDJF.index.hour,obsDJF.index.month]).describe()
```

```python
hour_month.head()
```

```python
hour_month.xs(1,level=0).head()
```

```python
hour_month.xs(1,level=1)[['mean','min','max']].plot(title='January NetRad')
```

```python
plt.show()
```

Gridspec
====

```python
import matplotlib.gridspec as gridspec
fig = plt.figure(figsize=(11,3.5))
gs = gridspec.GridSpec(2,2, width_ratios=[1,3], height_ratios=[1,3])
gs.update(wspace=0.16,hspace=0.1)
```

```python
ax1 = plt.subplot(gs[0,0])
ax1.text(0.5,0.5,'ax1')

ax2 = plt.subplot(gs[1,0])
ax2.text(0.5,0.5,'ax2')

ax3 = plt.subplot(gs[0:,1])
ax3.text(0.5,0.5,'ax3')

ax1.tick_params(labelbottom='off')
```

```python
plt.show()
```

```python
fig = plt.figure(figsize=(11,3.5))
gs = gridspec.GridSpec(2,2, width_ratios=[2,1], height_ratios=[1,1])
gs.update(wspace=0.16,hspace=0.3)
ax1 = plt.subplot(gs[0,0])
ax2 = plt.subplot(gs[1,0])
ax3 = plt.subplot(gs[0:,1])
ax1.tick_params(labelbottom='off')
hour_month.xs(1,level=1)[['mean','min','max']].plot(title='January NetRad',ax=ax1)
ax1.set_ylim((-200,900))
hour_month.xs(2,level=1)[['mean','min','max']].plot(title='February NetRad',ax=ax2)
ax2.set_ylim((-200,900))
plt.show()
```
